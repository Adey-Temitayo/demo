```python
df = pd.read_csv("student.csv")
df.head()
```

```python
df["total_score"]=100 * (df["math score"] + df["reading score"] + df["writing score"]) /300
df.head()
```

```python
df["total_score"]=round(100 * (df["math score"] + df["reading score"] + df["writing score"]) /300,2)
df.head()
```

```python
def grade(x):
    if x >= 90 and x <= 100:
        return "A"
    elif x >= 80 and x < 90:
        return "B"
    elif x >= 70 and x < 80:
        return "C"
    elif x >= 60 and x < 70:
        return "D"
    else:
        return "F"
```

```python
df["grade"] = df["total_score"].apply(grade)
df["grade"] 
```

```python
df["grade"] = df["total_score"].apply(grade)
df.head()
```

```python
sns.countplot(df,x="grade")
```

```python
# used to measure percentile 
plt.figure(figsize=(10,5))
sns.boxplot(df,x="parental level of education",y="math score")
```
